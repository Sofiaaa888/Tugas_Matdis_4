```python
import matplotlib.pyplot as plt
import pandas as pd

# Batas n
N = 15

# Hitung deret harmonik H_n
H = [0.0]  # H_0 = 0
for k in range(1, N+1):
    H.append(H[-1] + 1.0/k)

# Buat tabel dengan kolom True/False
data = []
for i in range(1, N+1):
    status = H[i] < 2   # True kalau masih < 2
    data.append([i, H[i], status])

df = pd.DataFrame(data, columns=["n", "H_n", "Benar (<2)?"])
print(df)

# Cari simpangan pertama (H_n >= 2)
first_n_ge_2 = None
for i, val in enumerate(H):
    if val >= 2.0:
        first_n_ge_2 = i
        break

print("\nSimpangan pertama pada n =", first_n_ge_2, "dengan nilai H_n =", H[first_n_ge_2])

# Plot grafik
plt.figure(figsize=(8,5))
plt.plot(range(N+1), H, marker='o', markersize=4, label="H_n")
plt.axhline(2.0, color='r', linestyle='--', label="y=2")
plt.scatter([first_n_ge_2], [H[first_n_ge_2]], color='red')
plt.text(first_n_ge_2+0.3, H[first_n_ge_2],
         f"n={first_n_ge_2}, H_n={H[first_n_ge_2]:.3f}")
plt.xlabel("n")
plt.ylabel("H_n")
plt.title("Deret Harmonik vs Batas y=2")
plt.grid(True)
plt.legend()
plt.show()
```

Gambar dan tabel menunjukkan perkembangan nilai deret harmonik parsial *Hn* seiring pertambahan *n*. Pada awalnya, nilai *Hn* meningkat cukup cepat. Dari tabel terlihat bahwa hingga *n*=3, nilai *Hn* masih berada di bawah 2. Namun, ketika *n*=4, nilai *Hn*=2,0833 yang berarti deret untuk pertama kalinya melampaui batas *y*=2. Inilah yang disebut sebagai titik simpangan pertama.
Grafik memperlihatkan kurva *Hn* yang semakin menanjak, meskipun kenaikannya semakin melambat seiring pertambahan nnn. Garis horizontal merah pada *y*=2 menunjukkan batas yang dibandingkan dengan deret. Titik potong ditandai jelas pada *n*=4, menandakan momen ketika deret melampaui batas tersebut.
Secara keseluruhan, interpretasi ini memperlihatkan bahwa deret harmonik bersifat divergen, artinya nilainya akan terus bertambah tanpa batas meskipun kenaikannya semakin lambat. Hal ini tampak dari tren kurva yang terus naik melewati garis batas *y*=2, dan akan terus meningkat jika n diperbesar.
